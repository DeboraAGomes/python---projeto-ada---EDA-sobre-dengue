# **Atributos**

- Year (Anos) - Anos entre 2012 e 2021

- State - Podendo ser o pais, no caso, Brasil ou os estado brasileiros

- Mean_Temp - Temperatura média em graus Celsius

- Min_Temp - Temperatura mínima em graus Celsius

- Max_Temp - Temperatura máxima em graus Celsius

- Percipitation - Percipitação anual dada em milímetros

- Change_Tmp - Max_Tmp menos Min_Temp em Graus Celsius

- Cases - Número de Casos de Dengue Registrados

- Region - Macroregiões do Brasil: Norte (N), Nordeste (NE), Centro-Oeste (CO), Sudeste (SE), Sul (S)

- State_Area(km2) - Área do Estado / Área de uma Unidade Federativa Dada em Quilômetros ao Quadrado

- Population - Estimativa da População Total do país ou Unidade Federativa
